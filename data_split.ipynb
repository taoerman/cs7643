```python
data = np.load('./dataset/train.npy', allow_pickle=True).item()
```

```python
data
```

```python
sequence_keys = list(data['sequences'].keys())
random.shuffle(sequence_keys)
```

```python
len(sequence_keys)
```

```python
train_keys = sequence_keys[:60]
val_keys = sequence_keys[60:]
```

```python
train_data = {
    'vocabulary': data['vocabulary'],
    'sequences': {k: data['sequences'][k] for k in train_keys}
}
val_data = {
    'vocabulary': data['vocabulary'],
    'sequences': {k: data['sequences'][k] for k in val_keys}
}
```

```python
np.save('dataset/train_split.npy', train_data)
np.save('dataset/val_split.npy', val_data)
```
